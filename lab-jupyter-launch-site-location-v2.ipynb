```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
spacex_df = pd.DataFrame({
    'Launch_Site': ['CCAFS LC-40', 'VAFB SLC-4E', 'KSC LC-39A', 'CCAFS SLC-40', 'CCAFS LC-40'],
    'Lat': [28.562302, 34.632834, 28.573255, 28.563197, 28.562302],
    'Long': [-80.577356, -120.610746, -80.646895, -80.577064, -80.577356],
    'class': [1, 0, 1, 1, 0]
})

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be 'green', if class=0, 'red'
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Display the updated DataFrame
spacex_df
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians
from folium import DivIcon

# Function to calculate the distance between two geographic coordinates (lat, lon)
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Initialize the map at the launch site (e.g., Kennedy Space Center)
launch_site_lat = 28.5721
launch_site_lon = -80.6480
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add MousePosition to map to show coordinates as the user moves the mouse
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Example coordinates for a city, railway, and highway (replace with actual POI coordinates)
city_lat = 28.5383  # Example city coordinates near Kennedy Space Center (Orlando)
city_lon = -81.3792
railway_lat = 28.5746  # Example railway coordinates near Kennedy Space Center
railway_lon = -80.6503
highway_lat = 28.5673  # Example highway coordinates near Kennedy Space Center
highway_lon = -80.5708

# Calculate distances between the launch site and the POIs
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Create and add markers for the POIs (city, railway, highway)
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM to City</b></div>'.format(distance_city),
    )
)
city_marker.add_to(site_map)

railway_marker = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM to Railway</b></div>'.format(distance_railway),
    )
)
railway_marker.add_to(site_map)

highway_marker = folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM to Highway</b></div>'.format(distance_highway),
    )
)
highway_marker.add_to(site_map)

# Add a marker for the launch site as well for reference
folium.Marker([launch_site_lat, launch_site_lon], popup="Launch Site").add_to(site_map)

# Create PolyLines to draw lines from launch site to each POI (city, railway, highway)
coordinates_city = [(launch_site_lat, launch_site_lon), (city_lat, city_lon)]
coordinates_railway = [(launch_site_lat, launch_site_lon), (railway_lat, railway_lon)]
coordinates_highway = [(launch_site_lat, launch_site_lon), (highway_lat, highway_lon)]

city_line = folium.PolyLine(locations=coordinates_city, weight=2, color='green').add_to(site_map)
railway_line = folium.PolyLine(locations=coordinates_railway, weight=2, color='blue').add_to(site_map)
highway_line = folium.PolyLine(locations=coordinates_highway, weight=2, color='red').add_to(site_map)

# Display the map
site_map
```
